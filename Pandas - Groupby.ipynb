## Groupby :
- It is a very powerful tool for data analysis and data manipulation.
- Pandas groupby function is used to dataframe or series data based on certain criteria . 
- Using groupby we can split the data into groups based on one or more columns and we can apply a function or comdition to each group and then we combine the groups.
